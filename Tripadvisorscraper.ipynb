## Import Hotels file

```python
# file that holds hotel information (url, review no's, pages..etc)
accommfile = r'CairnsAccom.csv'

df = pd.read_csv(os.path.join(path, accommfile))
df
```

```python
# select hotel to scrape data from eg 0 = 1st hotel in list; 5 = 6th hotel in list

selected_hotel = 11
```

```python
# setup defaults for scraper
urlid = df['urlid'].iloc[selected_hotel]
file = df['filename'].iloc[selected_hotel]
num_page = df['Pages'].iloc[selected_hotel]-1
path_to_file = os.path.join(dir_name,file)
num_page, path_to_file
```

```python
url = (urlpre+urlid+urlsuf)
url
```

```python
# if you pass the inputs in the command line
if (len(sys.argv) == 4):
    path_to_file = sys.argv[1]
    num_page = int(sys.argv[2])
    url = sys.argv[3]
```

```python
# import the webdriver
driver = webdriver.Chrome()

driver.get(url)
```

```python
# setup blank arrays for webscapper
reviewDetails =[]
start_time = time.time()


for i in range(0, num_page):

    # expand the review 
    time.sleep(2)
  
    reviewheader = driver.find_elements("xpath",".//div[@class='sCZGP']")
    reviewdetails = driver.find_elements("xpath","//div[@data-reviewid]")
 
    for j in range(len(reviewdetails)):
        try:
            reviewer = reviewheader[j].find_element("xpath",".//div[contains(@class, 'cRVSd')]").text
        
        except: NoSuchElementException
         
        try:    
            reviewerdetails = reviewheader[j].find_element("xpath",".//div[contains(@class,'MziKN')]").text
            
        except: NoSuchElementException
        
        try:    
            title = reviewdetails[j].find_element("xpath",".//div[contains(@data-test-target, 'review-title')]").text
            
        except: NoSuchElementException
        
        try:
            review = reviewdetails[j].find_element("xpath",".//q[@class='QewHA H4 _a']").text.replace("\n", "  ")
            
        except: NoSuchElementException
            
        try:
            rating = reviewdetails[j].find_element("xpath",".//span[contains(@class, 'ui_bubble_rating bubble_')]").get_attribute("class").split("_")[3]
            
        except: NoSuchElementException
            
        try:
            staydate = reviewdetails[j].find_element("xpath",".//span[contains(@class, 'teHYY _R Me S4 H3')]").text
            
        except: NoSuchElementException   
            
        result = [rating, title, review,reviewer, reviewerdetails, staydate]
        
        
           
        # prepare list of results from above arrays 
        reviewDetails.append(result)
      
        


     
    try:   
        # change the page            
        driver.find_element("xpath",'.//a[@class="ui_button nav next primary "]').click()
    except: NoSuchElementException
  
    print('Processed page: %s' %i)
    print("")
    print('Progress:' "{0:.2f}%".format((i + 1) / num_page * 100))
    print("")
    print("--- %s minutes ---" % "{:.2f}".format((time.time() - start_time)/60))
    print('-----------')
          
driver.quit()
```

```python
# make dataframe from results
reviewDetails = pd.DataFrame(reviewDetails,columns =['rating', 'title', 'review','reviewer', 'reviewerdetails', 'staydate']) 
reviewDetails['hotel'] = df['code'].iloc[selected_hotel]
reviewDetails
```

```python
# create csv files of results
reviewDetails.to_csv(path_to_file)
```
